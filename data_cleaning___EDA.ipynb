```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset= pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

DATA cleaning && EDA

```python
## missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis = 1)]
```

the data set is converted into two sets based on regions from 122 index , we can ,make a new column based on the Region
1 : "Bejaia region Dataset"
2 : "Sidi-Abbes Region Dataset"
Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
## chnage to int
df[['Region']] = df[['Region']].astype(int)
```

```python
df
```

```python
df.isnull().sum()
```

```python
## Removing th e null valus
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum().sum()
```

```python
df.loc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
## remove 122 row
```

```python
df.loc[[122]]
```

```python
df.columns
```

Here same columns has extra space like **'Classes  '**

```python
## fix spaces in columns name s
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
## Chnages the requied columns as integer data typ e
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## changing the inf of others columns as float data type
```

```python
objects = [featurs for featurs in df.columns if df[featurs].dtypes=='O']
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## let ave the cleaned data set
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
